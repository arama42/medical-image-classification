```python
import matplotlib.pyplot as plt
#!pip install opencv-python
import cv2
import numpy as np
import glob
```

```python
mis = './Misclassifications/PNEUMONIA'
```

```python
cor = './CorrectClassifications/PNEUMONIA'
```

```python
#Taking 50 misclassified images(pneumonia) to analyze the distribution
mis_imgs = []
imgs = glob.glob(mis + '/' + '*')
i = 0
for img in imgs:
    i+=1
    if i == 49:
        break
    img = cv2.imread(img)
    img_resized = cv2.resize(img, (299,299)) 
    mis_imgs.append(img_resized)
```

```python
#Taking 50 correct images(pneumonia) to analyze the distribution
corr_imgs = []
imgs = glob.glob(cor + '/' + '*')
i = 0
for img in imgs:
    i+=1
    if i == 49:
        break
    img = cv2.imread(img)
    img_resized = cv2.resize(img, (299,299)) 
    corr_imgs.append(img_resized)
```

```python
histr = cv2.calcHist(mis_imgs, [2], None, [256], [0, 256])
plt.plot(histr, color='r')
plt.xlim([0, 255])
```

```python
histg = cv2.calcHist(corr_imgs, [0], None, [256], [0, 256])
plt.plot(histg, color='g')
plt.xlim([0, 255])
```
